```python
import inspect
print("the class heirarchy for built-in")
inspect.getclasstree(inspect.getmro(Exception))
def Classtree(cls,indent=0):
    print('_'*indent, cls.__name__)
    for subcls in cls.__subclasses__():
        Classtree(subcls, indent+3)
Classtree(Exception)        
```

```python
"""QTS.4 - Lookup error is used when something is not found.key error and index error
           are the two of it's kind."""
try:
    l=[1,2,3,4,5,6,7]
    print(l[9])
except LookupError as e:
    print("index error",e)
    
```

```python
"""example of index error"""
try:
    l=[1,2,3,4,5,6,7]
    print(l[9])
except IndexError as e:
    print("index error",e)
    
```

```python
"""example of key error"""
try:
    d={"key1":1,"key2":"sudh","l":[1,1,2,3,4],"t":(2,3,4)}
    print(d["key3"])
except KeyError as e:
    print("key error",e)
```
